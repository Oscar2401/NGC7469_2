#### Flags:

```python
def flag_kau (x,y):
    sfg_flag_kau = np.zeros(np.shape(x))
    y_kau_us = np.zeros(np.shape(x))
    #Regiones de Formacion estelar (abajo de kuffmann) 
    for i in range(len(x)):
        y_kau_us[i] = np.divide(0.61,y[i]-0.05)+1.3
        if x[i] <= y_kau_us[i]:
            sfg_flag_kau[i] = 1.0
        else:
            sfg_flag_kau[i] = -999.00
    
    return sfg_flag_kau

def flag_kew(x,y):
    sfg_flag_kew = np.zeros(np.shape(x))
    y_kew_us = np.zeros(np.shape(x))
    #Regiones bajo Kewley
    for i in range(len(x)):
        y_kew_us[i] = np.divide(0.61,y[i]-0.47)+1.19
        if x[i] <= y_kew_us[i]:
            sfg_flag_kew[i] = 1.0
        else:
            sfg_flag_kew[i] = -999.00
    
    return sfg_flag_kew
```

# Diagrama BPT

The BPT diagrams (named after "Baldwin, Phillips & Telervich") are a set of nebular emission line diagrams used to distinguish the ionization mechanism of nebular gas (in this case, star formation and AGN). The most famous version consists of [NII]6584/Ha versus [OIII]5007/Hb  (the BPT-NII diagram; Fig. 5 of Baldwin et al 1981).  The next two more commonly-used BPT diagnostics are [SII]6717,6731/Ha versus [OIII]5007/Hb (BPT-SII) and [OI]6300/Ha versus [OIII]5007/Hb (BPT-OI).  These diagrams were studied in numerous work and the dividing lines have been developed and adapted as a function of the ionization models and/or observations available (e.g., Veilleux & Osterbrock 1987; Osterbrock 1989; Kewley et al 2001; Kauffmann et al 2003; Kewley et al 2006; Stasinska et al 2006; Kewley et al 2013a,b).  

```python
def div(l_x,l_y,x,y):
    
    #Hacemos los flageos con el diagrama BPT para las calibracioens 
    sfg_flag_kau =  flag_kau (l_x,l_y)
    sfg_flag_kew =  flag_kew (l_x,l_y)
    
    #Regiones de Formacion estelar (abajo de kuffmann)
    rest01 = x[np.where(sfg_flag_kau == 1.0)]
    log_rest01 = y[np.where(sfg_flag_kau == 1.0)]

    #Región entre kuffmann y Kewley
    doublet01 = np.logical_and(sfg_flag_kew == 1.0, sfg_flag_kau != 1.0)
    rest02 = x[np.where(doublet01)]
    log_rest02 = y[np.where(doublet01)]

    #Regiones de AGN (arriba de Kewley)
    rest03 = x[np.where(sfg_flag_kew != 1.0)]
    log_rest03 = y[np.where(sfg_flag_kew != 1.0)]
    
    return [rest01,log_rest01,rest02,log_rest02,rest03,log_rest03]
```

# Ajuste O3N2

#### The O3N2 index

The O3N2 index depends on 2 fractions of emission lines and was introduced by Alloin et al. (1979)

\begin{equation}
    O3N2 = log \left( \frac{[OIII]\lambda 5007}{H\beta} \times \frac{H\alpha}{[NII]\lambda 6584 } \right)
\end{equation}

```python
O3N2_6584_5007 = l_OIII_5007 - l_NII_6584
```

We calculate the abundance of oxygen using O3N2 Te-calibration (R.A. Marino et al. 2013):

\begin{equation}
    12 + log(O/H) = 8.553[\pm0.012]-0.214[\pm0.012]\times O3N2
\end{equation}


```python
l_HO_6584_5007 = 8.533 - 0.214*O3N2_6584_5007
```

```python
rest_O3N2 = div(l_OIII_5007,l_NII_6584,l_HO_6584_5007,O3N2_6584_5007)
```

#### We plot oxygen abundance versus O3N2 index 

```python
x1_fit_sun = np.arange(-1.1,1.7,0.00001)
y1_fit_sun = 8.533 - 0.214*x1_fit_sun
y1_fit_sun01 = (8.533 - 0.214*x1_fit_sun)+0.36 
y1_fit_sun02 = (8.533 - 0.214*x1_fit_sun)-0.36
```

```python
plt.figure(figsize=(8,8))
plt.plot(rest_O3N2[1],rest_O3N2[0],'b.')
plt.plot(rest_O3N2[3],rest_O3N2[2],'g.')
plt.plot(rest_O3N2[5],rest_O3N2[4],'r.')
plt.plot(x1_fit_sun,y1_fit_sun,'k-')
plt.plot(x1_fit_sun,y1_fit_sun01,'k--')
plt.plot(x1_fit_sun,y1_fit_sun02,'k--')
plt.xlabel('$\log[([OIII]\lambda5007 / H \\beta )/([NII]\lambda6584/ H\\alpha )]$',fontsize = 18)
plt.ylabel('$12 + \log(H/O)$',fontsize = 18)
plt.xlim(-2.0,3.0)
plt.ylim(7.5,9.2)
plt.title('O3N2 Calibration', fontsize = 18)
```

#### Indice N2 

El indice N2 ha sido estudiado por varios grupos, entre ellos Storchi-Bergmann et al. (1994), van Zee et al. (1998), Raimann et al. (2000), Denicoló et al. (2002), etc.

Este parametro es muy útil al calcular la abundancia de oxigeno por diferentes razones: 
(1) Es muy sensible al contenido de metal de una nevulosa.
(2) No sufre problemas de correción de enrojecimiento o calibración de flujo debido a la longitud de onda cercana de estas líneas.
(3) Puede detectarse con los espectógrafos de infrarojo cercano de nueva generación. 

We calculate the N2 index for each flow:
\begin{equation}
    N2 = log \left( \frac{[NII]\lambda 6584}{H\alpha} \right)
\end{equation}

```python
N2_6584 = l_NII_6584
```

We calculate oxygen abundance using O3N2 Te-calibration (R.A. Marino et al. 2013):

\begin{equation}
    12 + log(O/H) = 8.743[\pm0.027]+0.462[\pm0.024]\times N2
\end{equation}

```python
#l_HO_6584_N2 = 8.743 + 0.462*N2_6584
l_HO_6584_N2 = 8.533 + 0.214*O3N2_6584_5007
```

```python
rest_N2 = div(l_OIII_5007,l_NII_6584,l_HO_6584_N2, N2_6584)
```

# Prueba de calibración NS 
L.S. Pilyugin, L. Mattson (2011) Abundance determination in HII regions from spectra without the $[OII]\lambda 3727 + \lambda 3729$ line.

Usaremos la calibración sugerida por Pilyugin et al. 2011 para la determinación de abundancia de oxígeno y nitrógeno en regiones HII en las cuales las lineas [OII]$\lambda3727+\lambda3729$ no son faciles de conseguir

```python
#Coeficientes:
R_3 = np.divide(OIII_4959_flx+OIII_5007_flx,H_beta_flx)
N_2 = np.divide(NII_6548_flx+NII_6584_flx,H_beta_flx)
S_2 = np.divide(SII_6717_flux+SII_6731_flux,H_beta_flx)

Z_O = np.zeros(np.shape(R_3))
Z_N = np.zeros(np.shape(R_3))

#calibración:
for i in range(len(R_3)):
    if np.log10(N_2[i])>-0.1:
        Z_O[i] = 8.454-0.216*np.log10(R_3[i])-0.362*np.log10(S_2[i])-0.101*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 7.414-0.383*np.log10(R_3[i])+0.119*np.log10(S_2[i])+0.988*np.log10(N_2[i]/S_2[i])
    if np.log10(N_2[i])<-0.1 and np.log10(N_2[i]/S_2[i])>-0.25:
        Z_O[i] = 8.456+0.082*np.log10(R_3[i])+0.391*np.log10(N_2[i])+0.290*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 7.250+0.078*np.log10(R_3[i])+0.529*np.log10(N_2[i])+0.906*np.log10(N_2[i]/S_2[i])
    if np.log10(N_2[i])<-0.1 and np.log10(N_2[i]/S_2[i])<-0.25:
        Z_O[i] = 7.881+0.929*np.log10(R_3[i])+0.650*np.log10(N_2[i])+0.025*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 6.599+0.888*np.log10(R_3[i])+0.663*np.log10(N_2[i])+0.371*np.log10(N_2[i]/S_2[i])
        

N1 = div(l_OIII_5007,l_NII_6584,Z_O,Z_N)
#N2 = div(l_OIII_5007,l_NII_6584,Z_N,O3N2_6584_5007)
```

```python
### centro x=27, y=38
dis = np.zeros(62*62)
for i in range(62):
    for j in range(62):
        dis[62*i+j] = math.sqrt((27-i)**2 + (38-j)**2)
        dis[62*i+j] = math.tan(dis[62*i+j]*0.2*4.85e-6)*60000
        if imagen_OH[i][j] == 7.59:
            np.delete(dis,62*i+j)
a = Z_O
b = Z_N
            
for i in range(62):
    for j in range(62):
        if imagen_OH[i][j] == 7.59:
            np.delete(a,62*i+j)
abu_OH = np.reshape(a, (1,np.product(a.shape)))

for i in range(62):
    for j in range(62):
        if imagen_NH[i][j] == 7.08:
            np.delete(b,62*i+j,0)
abu_N = np.reshape(b, (1,np.product(b.shape)))
```

```python
def lin_fit(x,y): 
    
    gradient, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    mn=np.min(x)-1
    mx=np.max(x)+1
    x1 = np.linspace(mn,mx,500)
    y1 = gradient*x1 + intercept
    
    return [x1, y1, gradient, intercept, r_value, p_value, std_err]
```

```python
def fit_2(x,y,part):

    bin_i = np.min(x)
    binn = (np.max(x)-np.min(x))/part

    abu = []
    dis = []
    
    for k in range(part):
        c = []
        d = []
        bin_f = (k+1)*binn
        for i in range(62):
            for j in range(62):
                if bin_i <= x[62*i+j] <= bin_f :
                    c.append(y[62*i+j])
                    d.append(x[62*i+j])
        abu.append(((np.max(c)-np.min(c))/2 ) + np.min(c))
        dis.append(np.mean(d))
        bin_i = bin_f
     
    #a = 0
    #for l in range(len(abu)) :
    #    if np.isnan(abu[l-a]) == True :
    #        abu = np.delete(abu,l-a)
    #        abu = np.delete(dis,l-a)
    #        a = a + 1
    
    return abu,dis
    
```

# Electron Densities

To calculate electron densities, we use the prescription employed in Sanders et al. (2016)

\begin{equation} 
n_e (R) = \frac{cR-ab}{a-R}
\end{equation}

where the flux ratio $R=f([S ii]λ6716)/f([S ii]λ6731])$,and $a=0.4315$, $b=2107$ and $c=627.1$

```python
R = np.divide(SII_6717_flux,SII_6731_flux)
```

```python
n_e = np.divide(627.1*R-(0.4315*2107),0.4315-R)
```

```python
map_n_e_1 = np.zeros((62, 62))
map_n_e_2 = np.zeros((62, 62))

for i in range(62):
    for j in range(62):
        
        if 0.7 < R[62*i+j] < 1.45:
            map_n_e_1[i][j] =  n_e[62*i+j]
            map_n_e_2[i][j] =  n_e[62*i+j]
        
        if 0.7 > R[62*i+j] or  1.45 < R[62*i+j]:
            map_n_e_1[i][j] = n_e[62*i+j] #- np.min(n_e)
            map_n_e_2[i][j] = -0.0
        
        if map_n_e_2[i][j] >= 800:
            map_n_e_2[i][j] = 0.0
```

```python
plt.rcParams['image.cmap'] = 'gist_heat'
```

```python
#Cargo datos para trabajar:
OIII = np.loadtxt('OSCARHbOIII49595007-55.dat')
OI = np.loadtxt('OSCAROI6300-55.dat')
NII = np.loadtxt('OSCARHaNII65486584-55.dat')
SII = np.loadtxt('OSCARSII67176731-55.dat')
```

```python
H_beta_flx = OIII[:,3]
OIII_4959_flx = OIII[:,7]
OIII_5007_flx = OIII[:,11]
OI_6300_flx = OI[:,3]
H_alpha_flx = NII[:,3]
NII_6548_flx = NII[:,7]
NII_6584_flx = NII[:,11]
SII_6717_flx = SII[:,3]
SII_6731_flx = SII[:,7]
```

```python
log_O3Hb = np.log10(np.divide(OIII_5007_flx,H_beta_flx))
log_O1Ha = np.log10(np.divide(OI_6300_flx,H_alpha_flx))
log_N2Ha = np.log10(np.divide(NII_6584_flx,H_alpha_flx))
log_S2Ha = np.log10(np.divide(SII_6717_flx+SII_6731_flx,H_alpha_flx))
```

```python
#Coeficientes:
R_3 = np.divide(OIII_4959_flx+OIII_5007_flx,H_beta_flx)
N_2 = np.divide(NII_6548_flx+NII_6584_flx,H_beta_flx)
S_2 = np.divide(SII_6717_flx + SII_6731_flx,H_beta_flx)
Z_O = np.zeros(np.shape(R_3))
Z_N = np.zeros(np.shape(R_3))
```

```python
#calibración:
for i in range(len(R_3)):
    if np.log10(N_2[i])>-0.1:
        Z_O[i] = 8.454-0.216*np.log10(R_3[i])-0.362*np.log10(S_2[i])-0.101*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 7.414-0.383*np.log10(R_3[i])+0.119*np.log10(S_2[i])+0.988*np.log10(N_2[i]/S_2[i])
    if np.log10(N_2[i])<-0.1 and np.log10(N_2[i]/S_2[i])>-0.25:
        Z_O[i] = 8.456+0.082*np.log10(R_3[i])+0.391*np.log10(N_2[i])+0.290*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 7.250+0.078*np.log10(R_3[i])+0.529*np.log10(N_2[i])+0.906*np.log10(N_2[i]/S_2[i])
    if np.log10(N_2[i])<-0.1 and np.log10(N_2[i]/S_2[i])<-0.25:
        Z_O[i] = 7.881+0.929*np.log10(R_3[i])+0.650*np.log10(N_2[i])+0.025*np.log10(N_2[i]/S_2[i])
        Z_N[i] = 6.599+0.888*np.log10(R_3[i])+0.663*np.log10(N_2[i])+0.371*np.log10(N_2[i]/S_2[i])
```

```python
N1 = div(log_O3Hb,log_N2Ha,Z_O,Z_N)
```

```python
plt.figure()
plt.plot(N1[0],N1[1],'b.')
plt.plot(N1[2],N1[3],'g.')
plt.plot(N1[4],N1[5],'r.')
plt.xlabel('12 + $\log$(O/H)',fontsize = 18)
plt.ylabel('12 + $\log$(N/H)',fontsize = 18)
```
